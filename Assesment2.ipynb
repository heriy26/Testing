```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.preprocessing import MinMaxScaler
import statsmodels.api as sm
```

```python
df = pd.read_excel('gdp.xlsx')
df.head(10)
```

```python
df1 = df.interpolate(method ='linear', limit_direction ='forward')
df1.head(10)
```

```python
scaler = MinMaxScaler().fit(df1)

hasil = scaler.transform(df1)
print(hasil)
```

```python
corr = df1.corr()
round(corr, 3) 
```

```python
fig, ax = plt.subplots(figsize=(15, 5))
sns.heatmap(df1.corr(), center=0, cmap='YlOrBr', annot=True)
```

Dari analisis korelasi diatas dapat diambil kesimpulan bahwa faktor yang paling mempengaruhi GDP adalah non life premi dengan nilai r positif 0.67, dengan kata lain nilai r mendekakati nilai positif 1, yang artinya gdp dan non life premi memiliki korelasi searah.

```python
endog = df1['gdp']
model =sm.tsa.SARIMAX(endog, trend='c')
res = model.fit()

print(res.summary())
```

```python
pred = res.get_forecast(5)

print(pred.summary_frame(alpha=0.05))
```
